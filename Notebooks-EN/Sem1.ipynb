### Weighted mean

$$ \overline{x} = \dfrac{\sum_{i = 1}^N w_i x_i}{\sum_{i = 1}^N w_i}$$

Let's create some kind of list of weights. Learn how the `range()` method works

```python
w = []
for item in range(60):
    w.append((item + 1)/60)
```

```python
print(w)
```

```python
len(w)
```

Find the weighted average using the `np.average()` method, passing column `al` and weights `w` as arguments to the method.

```python
np.average(al, weights= w)
```

Let's calculate percentiles. Percentiles are the value that a given random variable does not exceed with a fixed probability.

```python
#Note that we use the pct = True flag

df['percentile'] = df['Al'].rank(pct = True)
```

```python
df.head()
```

```python
#Let's create a new dataframe with only al and percentile values

dfControl = df[['Al', 'percentile']]
```

```python
dfControl.head()
```

Now we need to sort the table in ascending order  

```python
dfControl = dfControl.sort_values(by = 'percentile', ascending=True)
```

```python
dfControl.head()
```

```python
dfControl.tail()
```

We need to plot the normal paper. To do this, we need to take the values from the percentile column, and call the inverse of the distribution function.

```python
dfControl['res'] = scipy.stats.norm.ppf(dfControl['percentile'])
```

```python
dfControl.head()
```

Now let's make a dot plot of what we get.

```python
plt.pyplot.plot(dfControl['Al'], dfControl['res'], 'o')
```

Finally, let's construct a continuous graph. 

```python
plt.pyplot.plot(dfControl['Al'], dfControl['res'])
```

What do we see? That the plotting of the graph is almost linear. So we can conclude that the distribution is normal!
